```python
a_0 = 1
G_N = 6.67*np.power(10., -11.)

def t_f(xi, N_e):
    return N_e*np.sqrt((3*G_N*(1+xi*np.exp(N_e))*(1+2*xi))/(np.power(1+xi, 2.)))

def G_0(xi, N_e):
    return G_N*(1+xi*np.exp(N_e))

def Lambda_0(xi, N_e):
    return np.power(1+xi, 2.)/(G_N*(1+xi*np.exp(N_e))*(1+2*xi))

def a(xi, N_e, t):
    return a_0*np.exp(t*np.sqrt(Lambda_0(xi, N_e)/3))

def G(xi, N_e, t):
    return G_0(xi, N_e)/(1+xi*a(xi, N_e, t))

def Lambda(xi, N_e, t):
    return Lambda_0(xi, N_e)*(1+2*xi*a(xi, N_e, t))/(1+xi*a(xi, N_e, t))

Px = 100
PN = 500

log_xi = np.arange(-Px, Px+0.1, 0.1)
xim = np.zeros(2*10*Px, dtype=np.float64)
Nem = np.arange(0, PN+0.25, 0.25)
xi_validos = np.zeros(184015, dtype=np.float64)
Ne_validos = np.zeros(184015, dtype=np.float64)

for i in range(2*10*Px):
    xim[i] = np.exp(log_xi[i])

k = 0

for i in range(2*10*Px):
    for j in range(2*10*Px):
        tf = t_f(xim[i], Nem[j])
        if Lambda(xim[i], Nem[j], tf)*G(xim[i], Nem[j], tf) <= np.power(10., -61.):
            if np.power(10., -71.) <= Lambda(xim[i], Nem[j], tf)*G(xim[i], Nem[j], tf):
                if (1+xim[i]*np.exp(Nem[j]))*(1+2*xim[i]) > 0:
                    if Nem[j] >= 60:
                        xi_validos[k] = log_xi[i]
                        Ne_validos[k] = Nem[j]
                        k = k + 1
```
